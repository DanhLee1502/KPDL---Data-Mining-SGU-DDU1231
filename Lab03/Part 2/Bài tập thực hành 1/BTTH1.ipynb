## Chuẩn bị dữ liệu cho trực quan hóa dữ liệu

```python
import numpy as np
import pandas as pd
data_path = r"C:\Users\MSI GAMING\Documents\Khai phá dữ liệu\Lab03_Lê Thành Danh_3123580005\Phần 2\Bài tập thực hành 1\winequality-red.csv"
df = pd.read_csv(data_path)
```

```python
df.info()
print("\nCác giá trị bị thiếu:\n",df.isnull().sum())
```

```python
dfsort = df.sort_values('quality', ascending =
False)
dfsort.head()
```

## Trực quan hóa dữ liệu với thư viện Matplotlib

```python
import matplotlib.pyplot as plt
wine_quality_avg = dfsort.groupby('quality').mean().reset_index()
#case 1
x = wine_quality_avg['quality']
y_alcohol = wine_quality_avg['alcohol']
y_acidity = wine_quality_avg['volatile acidity']
plt.figure(figsize=(12, 7))
plt.bar(x, y_alcohol)
plt.show()
```

```python
#Case 2
plt.figure(figsize=(12, 7))
plt.bar(x, y_alcohol, color='darkred')
plt.title('Nồng độ cồn trung bình theo chất lượng rượu', fontsize=18)
plt.xlabel('Chất lượng rượu', fontsize=15)
plt.ylabel('Nồng độ cồn trung bình (%)', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### Biểu đồ nồng độ cồn và độ axit dễ bay hơi trung bình theo chất lượng

```python
#Case 3
# --- Biểu đồ thứ nhất (Bên trái) ---
plt.subplot(1, 2, 1)
plt.bar(x, y_alcohol, color='darkred')
plt.xlabel('Chất lượng rượu', fontsize=7)
plt.ylabel('Nồng độ cồn trung bình (%)', fontsize=7)
plt.title('Nồng độ cồn trung bình theo chất lượng', fontsize=7)
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
# --- Biểu đồ thứ hai (Bên phải) ---
plt.subplot(1, 2, 2)
plt.bar(x, y_acidity, color='skyblue')
plt.xlabel('Chất lượng rượu', fontsize=7)
plt.ylabel('Độ axit dễ bay hơi trung bình', fontsize=7)
plt.title('Độ axit dễ bay hơi trung bình theo chất lượng', fontsize=7)
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.tight_layout(pad=3.0)
```

## Trực quan hóa dữ liệu với thư viện Seaborn

```python
import seaborn as sns
#Case 1
plt.figure(figsize=(12, 7))
sns.barplot(data=wine_quality_avg, x='quality', y='alcohol')
data = dfsort[0:10]
plt.title('Nồng độ cồn theo chất lượng')
plt.savefig('wine_seaborn_case1_basic.png')
plt.show() 
```

```python
#Case 2
plt.figure(figsize=(12, 7))
ax_case2 = sns.barplot(data=wine_quality_avg, x='quality', y='alcohol', palette='viridis')
ax_case2.set_xlabel('Chất lượng rượu', fontsize=15)
ax_case2.set_ylabel('Nồng độ cồn trung bình (%)', fontsize=15)
ax_case2.set_title('Case 2: Biểu đồ nâng cao', fontsize=20)
plt.show()
```

```python
#Case 3
fig, ax = plt.subplots(1, 2, figsize=(9,5))

ax1 = sns.barplot(data=wine_quality_avg, x='quality', y='alcohol', ax=ax[0], palette='Reds_r')
ax1.set_xlabel('Chất lượng rượu')
ax1.set_ylabel('Nồng độ cồn trung bình (%)')
ax1.set_title('Nồng độ cồn trung bình')

ax2 = sns.barplot(data=wine_quality_avg, x='quality', y='volatile acidity', ax=ax[1], palette='Blues_r')
ax2.set_xlabel('Chất lượng rượu')
ax2.set_ylabel('Độ axit dễ bay hơi trung bình')
ax2.set_title('Độ axit dễ bay hơi trung bình')
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```
