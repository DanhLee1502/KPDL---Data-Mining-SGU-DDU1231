```python
unit_price_col = df["UnitPrice"]
# Tính toán các giá trị
data_mean = np.mean(unit_price_col)
data_median = np.median(unit_price_col)
data_mode_result = stats.mode(unit_price_col)
data_variance = np.var(unit_price_col)
data_sd = np.std(unit_price_col)
data_range = np.max(unit_price_col) - np.min(unit_price_col)
data_percentile_60 = np.percentile(unit_price_col, 60)
data_quartile_75 = np.quantile(unit_price_col, 0.75)
data_iqr = stats.iqr(unit_price_col)
```

```python
print("\n[Nhiệm vụ 1: Kết quả thống kê mô tả]")
print(f" - Mean (Trung bình): {data_mean:.2f}")
print(f" - Median (Trung vị): {data_median:.2f}")

# --- Xử lý Mode an toàn ---
mode_value = data_mode_result.mode
if isinstance(mode_value, (list, np.ndarray, pd.Series)):
    if len(mode_value) > 0:
        print(f" - Mode (Yếu vị): {mode_value[0]}")
    else:
        print(" - Mode (Yếu vị): Không có giá trị mode duy nhất.")
else:
    print(f" - Mode (Yếu vị): {mode_value}")

# --- Các chỉ số thống kê khác ---
print(f" - Variance (Phương sai): {data_variance:.2f}")
print(f" - Standard Deviation (Độ lệch chuẩn): {data_sd:.2f}")
print(f" - Range (Phạm vi): {data_range}")
print(f" - 60th Percentile: {data_percentile_60:.2f}")
print(f" - 75th Quartile (Q3): {data_quartile_75:.2f}")
print(f" - Interquartile Range (IQR): {data_iqr:.2f}")
```

## Trực quan hóa dữ liệu với Matplotlib

```python
# Chỉ giữ lại các giao dịch có số lượng lớn hơn 0
df = df[df['Quantity'] > 0]
# Tạo cột doanh thu TotalPrice
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Tính tổng doanh thu theo từng quốc gia
country_sales = df.groupby('Country')['TotalPrice'].sum().reset_index()
# Sắp xếp và lấy top 10 quốc gia (bỏ qua UK)
top_countries = country_sales[country_sales['Country'] != 'United Kingdom'].nlargest(10, 'TotalPrice')


x_countries = top_countries['Country']
y_sales = top_countries['TotalPrice']

#Case 1
plt.figure(figsize=(10, 6))
plt.bar(x_countries, y_sales)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
## Case 2: Nâng cao 1
plt.figure(figsize=(12, 7))
plt.bar(x_countries, y_sales, color='darkred')
plt.title('Top quốc gia có tổng doanh thu cao nhất (Trừ Mỹ)', fontsize=16)
plt.xlabel('Quốc gia', fontsize=12)
plt.ylabel('Tổng doanh thu', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
## Case 3: Nâng cao 2 (Subplots)
# Chuẩn bị dữ liệu về số lượng khách hàng
country_customers = df.groupby('Country')['CustomerID'].nunique().reset_index()
top_countries_customers = pd.merge(top_countries, country_customers, on='Country')
y_customers = top_countries_customers['CustomerID']

# Tạo subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle('Doanh số bán hàng và số lượng khách hàng ở các quốc gia (Trừ Mỹ)', fontsize=18)
# Biểu đồ doanh thu
ax1.bar(x_countries, y_sales, color='skyblue')
ax1.set_title('Doanh số bán hàng ở mỗi quốc gia')
ax1.set_ylabel('Doanh số bán hàng')
ax1.tick_params(axis='x', rotation=60)
# Biểu đồ khách hàng
ax2.bar(x_countries, y_customers, color='darkred')
ax2.set_title('Số lượng khách hàng ở mỗi quốc gia')
ax2.set_ylabel('Số lượng khách hàng')
ax2.tick_params(axis='x', rotation=60)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
